```python
import csv
from urllib.request import urlopen, Request
import io

# Delimitador fixo para os arquivos CSV
# Agora a função retorna uma tupla: (headers, rows)
def load_csv(url, delimiter=',', encoding='utf-8'):
    req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
    with urlopen(req) as resp:
        text = resp.read().decode(encoding, errors='replace')
    f = io.StringIO(text)
    reader = csv.DictReader(f, delimiter=delimiter)
    return reader.fieldnames, list(reader)

orders_url = 'https://raw.githubusercontent.com/fiesc-junior-prado/mine_projeto_bloco_1/refs/heads/main/olist_orders_dataset.csv'
products_url = 'https://raw.githubusercontent.com/fiesc-junior-prado/mine_projeto_bloco_1/refs/heads/main/olist_products_dataset.csv'

print('Lendo orders...')
orders_headers, orders = load_csv(orders_url, delimiter=',')
print(f"{len(orders)} linhas em 'orders'.")
print('Cabeçalhos orders:', orders_headers)

print('Lendo products...')
products_headers, products = load_csv(products_url, delimiter=',')
print(f"{len(products)} linhas em 'products'.")
print('Cabeçalhos products:', products_headers)

# Mostrar primeiras 3 linhas de cada arquivo para conferência
print('\nPrimeiras 3 orders:')
for r in orders[:3]:
    print(r)

print('\nPrimeiros 3 products:')
for r in products[:3]:
    print(r)
```
